```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
df = pd.read_csv("./Fish.csv")
dico_especes = {
    "Bream": 1,
    "Roach": 2,
    "Whitefish": 3,
    "Parkki": 4,
    "Perch": 5,
    "Pike": 6,
    "Smelt": 7
}

df['espece'] = df['Species'].map(dico_especes)
df.head()
X_train, X_test, y_train, y_test = train_test_split(df.drop('espece', axis=1), df["espece"], test_size=0.20)
selected_features = ['Weight', 'Length1','Length2','Length3', 'Height', 'Width']
X_train = X_train[selected_features]
X_test = X_test[selected_features]

# 1 - create a linear regression model using LinearRegression from Scikit-learn
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
# Make predictions using the testing set
y_pred = regr.predict(X_test)
# The coefficients
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))

# 2 - plot the test set and the prediction of your model to compare them
# 3 - display the performance indicators of your model : R2
# Calculate R2 score
r2 = r2_score(y_test, y_pred)
print("R2 score:", r2)
if len(X_test) != len(y_pred):
    print("Warning: Array lengths differ. Plots might be inaccurate.")

else:
    # Scatter plot for actual vs predicted values
    plt.scatter(X_test['Length2'], y_test, color="blue", label="Actual Values")  
    plt.scatter(X_test['Length2'], y_pred, color="red", label="Predicted Values")
    plt.xlabel("Length2")
    plt.ylabel("Target Values")
    plt.legend()
    plt.text(0.5, 0.95, f'R2 Score: {r2:.2f}', horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

    # Plot R2 as a horizontal line
    plt.axhline(y=r2, color='green', linestyle='-', label=f'R2 = {r2:.2f}')
    
    plt.show()
```
